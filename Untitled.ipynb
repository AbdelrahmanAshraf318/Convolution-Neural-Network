```python
#rint(k.image_data_format()) (Channels_last)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
X_valid = X_valid.reshape(X_valid.shape[0], 28, 28, 1)
inpx = (28, 28, 1)

```

```python
def plot_image(X , y , index):
    classes = ["zero" , "one" , "two" , "three" , "four" , "five" , "six" , "seven" , "eight" , "nine"]
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]])
```

```python
plot_image(X_train , y_train , 0)
```

```python
def normalize(X):
    X = X.astype("float32")
    X = X/255
    return X
```

```python
X_train = normalize(X_train)
```

```python
X_train
```

```python
X_test = normalize(X_test)
```

```python
X_valid = normalize(X_valid)
```

<div style="color:black; border-radius: 5px; margin: 5px 0;font-family:'Source Sans Pro', sans-serif;">
 <h4 style="border-left: 4px solid #96D4D4; background: #D9EEE1; padding: 5px; font-family:'Source Sans Pro', sans-serif;">Code</h4>

```python
cnn = models.Sequential([
    # CNN layers
    layers.Conv2D(filters=32 , kernel_size=(3,3) , activation = 'relu' , input_shape=(28,28,1)), #Convolutional Layer to get the feature map
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(filters=64 , kernel_size=(3,3) , activation = 'relu'), #Convolutional Layer to get the feature map
    layers.MaxPooling2D((2,2)),
    
    #dense network
    layers.Flatten(),
    layers.Dense(64 , activation='relu'),
    layers.Dense(10 , activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=["accuracy"]
           )
```

```python
cnn.fit(X_train , y_train , epochs=10)
```

```python
cnn.evaluate(X_valid , y_valid)
```

```python
plot_image(X_valid , y_valid , 1)
```

```python
y_pred = cnn.predict(X_test)
y_classes = [np.argmax(element) for element in y_pred]
```
